```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
#loading data 
custmomer_data=pd.read_csv("Mall_Customers.csv")
custmomer_data.head()
```

```python
#represnting data of csv file
custmomer_data.shape
```

```python
#printing data type info
custmomer_data.info()
```

```python
#checking for missing value
custmomer_data.isnull().sum()
```

```python
#choosing annual income coulmn and spending score column
x=custmomer_data.iloc[:,[3,4]].values#it is column slcing#if we use .iloc[[3:4],:]means this raw slicing
print(x)
```

```python
#choosing number of cluster
#WCSS-> within cluster sum of square
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
#plot an elbow graph
sns.set()
plt.plot(range(1,11),wcss)
plt.title("The Elbow point graph")
plt.xlabel("number of cluster")
plt.ylabel("WCSS")
plt.show()
```

```python
#OPTIMUM NUMBER OF CLUSTER=5
#TRAINING THE K-means CLUSTER model
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=0)
y=kmeans.fit_predict(x)
print(y)
```

```python
#5 cluster=0,1,2,3,4
#visualizing all the cluster
plt.figure(figsize=(9,9))
plt.scatter(x[y==0,0],x[y==0,1],s=50,c="red",label='cluster_1')
plt.scatter(x[y==1,0],x[y==1,1],s=50,c="yellow",label='cluster_1')
plt.scatter(x[y==2,0],x[y==2,1],s=50,c="green",label='cluster_1')
plt.scatter(x[y==3,0],x[y==3,1],s=50,c="blue",label='cluster_1')
plt.scatter(x[y==4,0],x[y==4,1],s=50,c="brown",label='cluster_1')
#plot the centroids
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,label="centroids")
plt.title("Custmour Group")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```
